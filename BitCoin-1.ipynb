```python
# import the libraries
import pandas as pd
import numpy as np
```

```python
# load the data
from google.colab import files
uploaded = files.upload()
```

```python
# store the data into a variable
df = pd.read_csv('market-price.csv')

# show the first 7 rows of data
df.head(7)
```

```python
# remove the data column
df.drop(['Timestamp'], 1, inplace = True)
```

```python
# show the first seven rows of the new data set
df.head(7)
```

```python
# create a variable for predicting 'n' days out into the future
prediction_days = 30 #n = prediction days

#create another column shifted 'n' units up
df['Prediction'] = df[['market-price']].shift(-prediction_days)
```

```python
# show the first 7 rows of the new dataset
df.head(7)
```

```python
# show the last 7  rows of the new dataset
df.tail(7)
```

```python
# create the independent dataset X

# convert the data frame to a numpy array and drop the prediction column
X = np.array(df.drop(['Prediction'], 1))

# remove the last 'n' rows where 'n' is the prediction_days
X = X[:len(df)-prediction_days]

print(X)
```

```python
# create the dependent data set

# converte the dataframe to a numpy array
y = np.array(df['Prediction'])

# get all of the values except the last 'n' rows
y = y[:-prediction_days]
print(y)
```

```python
# split the data into 80% training and 20% testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# set the prediction_days_array equal to the last 30 rows from the original dataset
prediction_days_array = np.array(df.drop(['Prediction'], 1))[-prediction_days:]
print(prediction_days_array)
```

```python
from sklearn.svm import SVR

# create and train the support vector machine (regression) using radial basis function
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.00001)
svr_rbf.fit(x_train, y_train)
```

```python
# test the model
svr_rbf_confidence = svr_rbf.score(x_test, y_test)
print("svr_rbf accuracy: ", svr_rbf_confidence)
```

```python
# print the predicted values
svm_prediction = svr_rbf.predict(x_test)
print(svm_prediction)

print()

# print the actual values
print(y_test)
```

```python
# print the model predictions for the next 'n' = 30 days
svm_prediction = svr_rbf.predict(prediction_days_array)
print(svm_prediction)

print()

# print the actual price for bitcoin for the last 30 days
print(df.tail(prediction_days))
```
